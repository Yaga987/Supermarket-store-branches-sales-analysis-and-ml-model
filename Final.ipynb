```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.metrics import mean_absolute_error ,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler ,RobustScaler
from sklearn.svm import SVR,LinearSVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/Stores.csv',index_col=0)
df
```

```python
df.shape
```

```python
y = df['Store_Sales']
y = pd.DataFrame(y)
y
```

```python
x = df.drop(['Store_Sales'],axis=1)
x
```

```python
full_data = x
full_data['Sales'] = y
full_data
```

```python
data_Frame = pd.DataFrame(full_data)
data_Frame.to_csv('DP.csv',index=False)
```

```python
AV = AutoViz_Class()
piz = AV.AutoViz(
    'DP.csv',
    sep=',',
    chart_format = 'html', # 'server,'html,'png'
    verbose = 2
    # max_rows_analyzed = 200, col ve row sınırlama, verbose = 2 ile kayıtta edebilirsin resimleri
)
```

```python
x.shape,y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y.values, train_size=0.8)
```

```python
x_train = StandardScaler().fit_transform(x_train)
x_test = StandardScaler().fit_transform(x_test)
```

```python
model = SVR(kernel = "rbf", gamma = 1.5)
model.fit(x_train, y_train)
Y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, Y_pred)
r2s = r2_score(y_test,Y_pred)
print('SVRrbf')
print(f'mae: {mae}')
print(f'r2s: {r2s}')
```

```python
model = SVR(kernel = "linear", gamma = 1.5)
model.fit(x_train, y_train)
Y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, Y_pred)
r2s = r2_score(y_test,Y_pred)
print('SVRL')
print(f'mae: {mae}')
print(f'r2s: {r2s}')
```

```python
model = LinearSVR()
model.fit(x_train, y_train)
Y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, Y_pred)
r2s = r2_score(y_test,Y_pred)
print('LSVR')
print(f'mae: {mae}')
print(f'r2s: {r2s}')
```

```python
model = KNeighborsRegressor(n_neighbors=25)
model.fit(x_train, y_train)
Y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, Y_pred)
r2s = r2_score(y_test,Y_pred)
print('KNR')
print(f'mae: {mae}')
print(f'r2s: {r2s}')
```

```python
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
Y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, Y_pred)
r2s = r2_score(y_test,Y_pred)
print('DTR')
print(f'mae: {mae}')
print(f'r2s: {r2s}')
```

```python
model = RandomForestRegressor(n_estimators=25)
model.fit(x_train, y_train)
Y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, Y_pred)
r2s = r2_score(y_test,Y_pred)
print('RFR')
print(f'mae: {mae}')
print(f'r2s: {r2s}')
```
